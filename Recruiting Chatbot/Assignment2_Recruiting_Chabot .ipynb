```python
print("Hello, Welcome to our Job Portal.")
print('We will help you find your Dream Job.')
print('Help us with the following details.\n')

print('*************************************************************\n')

name = input('Enter your Full Name.')
city = input('Enter the city where you would preferrably want to work.')
gender = input('Enter your gender.')
age = input('Enter your age')

print('*************************************************************\n')

print('Following are the job roles.')
for i in job_roles.keys():
    print(i , job_roles[i])
job = job_roles[int(input('Select a job role from above list by entering the number.'))]

print('Following are the salaries.')
for i in salaries_lpa.keys():
    print(i , salaries_lpa[i])
salary_range = salaries_lpa[int(input('Select a salary range from above list by entering the number.'))]

print('*************************************************************\n')

print('------------------------USER DETAILS-------------------------\n')
print("Name                 : ",name)
print("Age                  : ",age)
print("Gender               : ",gender)
print("City                 : ",city)
print("Job Role             : ",job)
print("Salary Expectations  : ",salary_range)
print('*************************************************************\n')

print('Here are the job roles that match your requirements.')

for i in jobs.keys():
    if jobs[i]['Job_role'] == job and jobs[i]['Salary'] == salary_range:
        print('Company:  ', jobs[i]['Company'])
        print('Job role: ', jobs[i]['Job_role'])
        print('Salary:   ' , jobs[i]['Salary'])
        
print('*************************************************************\n')

phone = input("Please enter your Phone Number : ")
email = input("Please enter your email-id     : ")

details = name+","+age+","+city+","+job+","+salary_range+","+email+","+phone+"\n"
fd = open("Record.txt",'a')
fd.write(details)
fd.close()

print("Details saved!")
print('Thank you for connecting with us')
print('*************************************************************\n')
```
